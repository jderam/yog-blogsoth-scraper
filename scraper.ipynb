```python
import requests 
from bs4 import BeautifulSoup
```

```python
url = "http://yog-blogsoth.blogspot.com/"
```

```python
def getdata(url): 
    r = requests.get(url) 
    return r.text

def get_img_urls(url):
    txt = getdata(url)
    soup = BeautifulSoup(txt, 'html.parser')
    return [x["src"] for x in soup.find_all('img') if "/s400/" in x["src"]]

def save_image(monster_img):
    out_file = f"images/{monster_img.split('/')[-1].replace('%2B', '_')}"
    response = requests.get(monster_img)
    with open(out_file, "wb") as f:
        f.write(response.content)
    return
```

```python
main_page_data = getdata(url)
```

```python
soup = BeautifulSoup(main_page_data, 'html.parser') 
```

```python
monster_page_urls = [link.get("href") for link in soup.find_all("a") if link.get("href") is not None and "search/label" in link.get("href")]
```

```python
for mp in monster_page_urls:
    img_urls = get_img_urls(mp)
    for iu in img_urls:
        print(f"Getting {iu.split('/')[-1].replace('%2B', '_')}")
        img_url = iu.replace("/s400/", "/s1600/")
        save_image(img_url)
    
```
